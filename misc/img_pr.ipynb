### Reproduce the IMG/PR subset

```python
# download plasmid metadata from IMG/PR
# filename: IMGPR_plasmid_data.tsv (163 MB)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.25)

df = pd.read_csv('IMGPR_plasmid_data.tsv', sep='\t', header=0)
df.shape
```

```python
df.columns
```

```python
dfx = df.drop_duplicates(subset="ptu")

# total bases (in Millions) and average plasmid length (in bases)
# see Supplementary Table S3 in the paper
round(dfx.length.sum() / 1e6), round(dfx.length.mean())
```

```python
sns.histplot(data=dfx, x='length', log_scale=True, bins=30)
plt.xlabel('\nIMG/PR Plasmid Size (bp, log-scale)')
plt.ylabel('Number of Plasmids\n')
plt.savefig('img_pr_plasmid_size_log.png', dpi=800)
```
